## X Y split

```python
y_train = train_df["Activity"]
X_train = train_df.copy().drop(columns="Activity")
X_train = X_train.drop(columns="subject")
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_s = scaler.fit_transform(X_train)
```

```python
y_train = y_train.values
y_train
```

```python
X_train_s.shape, y_train.shape
```

## Training

```python
from sklearn.svm import SVC

svc_clf = SVC(C=100, kernel="rbf")
clf = svc_clf.fit(X_train_s, y_train)
```

## Prediction

```python
test_df = test_df.drop(columns="subject")
X_test = test_df.drop(columns="Activity")
y_test = test_df["Activity"]
```

```python
X_test_s = scaler.fit_transform(X_test)
```

```python
y_pred = clf.predict(X_test_s)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
